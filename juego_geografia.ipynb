# Preguntas y Respuestas


- Reglas del Juego de "Preguntas y Respuestas de GeografÃ­a":
    - El juego consiste en hacer preguntas de geografÃ­a y que el jugador adivine la respuesta correcta.

    - Cada pregunta tiene una respuesta Ãºnica y correcta.

    - El jugador tiene un nÃºmero limitado de intentos para responder correctamente a cada pregunta.

    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.


- Pistas para las alumnas:

    1. Comienza creando un diccionario que contenga preguntas de geografÃ­a como claves y respuestas como valores.

    2. Implementa una funciÃ³n para seleccionar una pregunta aleatoria del diccionario.

    3. Utiliza un bucle `while` para hacer preguntas de geografÃ­a hasta que el jugador responda un nÃºmero especÃ­fico de preguntas correctamente o alcance un nÃºmero mÃ¡ximo de respuestas incorrectas.

    4. Dentro del bucle, presenta la pregunta al jugador y permita que ingresen sus respuestas.

    5. Lleva un seguimiento del nÃºmero de preguntas realizadas y el nÃºmero de respuestas correctas.

    6. Determina las condiciones de victoria y derrota del juego y sal del bucle cuando se cumplan estas condiciones.

# JUEGO SIN EL EXTRA

```python
import random

# Diccionario con preguntas y respuestas
preguntas_respuestas = {
    'Â¿CuÃ¡l es la capital de Francia?': 'paris',
    'Â¿CuÃ¡l es el continente mÃ¡s grande?': 'asia',
    'Â¿QuÃ© paÃ­s tiene la mayor poblaciÃ³n del mundo?': 'india',
    'Â¿CuÃ¡l es el rÃ­o mÃ¡s largo del mundo?': 'amazonas',
    'Â¿En quÃ© paÃ­s se encuentra la Torre Eiffel?': 'francia',
    'Â¿CuÃ¡l es la capital de JapÃ³n?': 'tokio',
    'Â¿CuÃ¡l es el desierto caliente mÃ¡s grande del mundo?': 'sahara',
    'Â¿CuÃ¡l es la capital de Argentina?': 'buenosaires',
    'Â¿QuÃ© paÃ­s es conocido como la tierra de los canguros?': 'australia'
}

# Variables 
conteo_preguntas = 0
respuestas_correctas = 0
respuestas_incorrectas = 0

# Pasamos el diccionario a lista porque el mÃ©todo random choice se aplica a listas
preguntas_respuestas_lista = list(preguntas_respuestas)
# Cuando las condiciones se cumplan, se pararÃ¡ el juego con un break
while True: 
    # Generamos una pregunta aleatoria de la lista
    pregunta_random = random.choice(preguntas_respuestas_lista)
    # Llevamos la cuenta de las preguntas hechas, independientemente de si son correctas o no 
    conteo_preguntas += 1
    print(f"Esta es la pregunta nÃºmero {conteo_preguntas}")
    # Mostramos pregunta al usuario y recogemos respuesta
    respuesta_usuario = input(f"Responda la siguiente pregunta: {pregunta_random}")
    # Protegemos el formato de la respuesta del usuario para que coincida con la respuesta del diccionario
    respuesta_usuario = "".join(respuesta_usuario.split()).lower() 
    # Comparamos la respuesta del usuario con el value correspondiente a la pregunta (key) del diccionario
    if respuesta_usuario == preguntas_respuestas[pregunta_random]:
        respuestas_correctas += 1
        print(f"Has acertado! Llevas un total de {respuestas_correctas} respuestas correctas ğŸ¤©")
        if respuestas_correctas == 5:
            print(f"Enhorabuena, has ganado el juego! ğŸ‰")
            break
        else:
            print(f"A por la siguiente pregunta!ğŸ¤“")
            # Si adivina una pregunta y el juego sigue, sacar esa pregunta de la lista para que no se repita
            preguntas_respuestas_lista.remove(pregunta_random)
    else:
        respuestas_incorrectas += 1
        print(f"La respuesta no es correcta ğŸ™ Has fallado un total de {respuestas_incorrectas} veces.")
        if respuestas_incorrectas == 3:
            print(f"El juego ha terminado...ğŸ˜¢")
            break
```

```python
import random

# Diccionario con preguntas y respuestas
preguntas_respuestas = {
    'Â¿CuÃ¡l es la capital de Francia?': 'paris',
    'Â¿CuÃ¡l es el continente mÃ¡s grande?': 'asia',
    'Â¿QuÃ© paÃ­s tiene la mayor poblaciÃ³n del mundo?': 'india',
    'Â¿CuÃ¡l es el rÃ­o mÃ¡s largo del mundo?': 'amazonas',
    'Â¿En quÃ© paÃ­s se encuentra la Torre Eiffel?': 'francia',
    'Â¿CuÃ¡l es la capital de JapÃ³n?': 'tokio',
    'Â¿CuÃ¡l es el desierto caliente mÃ¡s grande del mundo?': 'sahara',
    'Â¿CuÃ¡l es la capital de Argentina?': 'buenosaires',
    'Â¿QuÃ© paÃ­s es conocido como la tierra de los canguros?': 'australia'
}

# Variables
intentos_pregunta = 2  # Intentos permitidos por cada pregunta
conteo_preguntas = 0
respuestas_correctas = 0
respuestas_incorrectas = 0

# Pasamos el diccionario a lista porque el mÃ©todo random choice se aplica a listas
preguntas_respuestas_lista = list(preguntas_respuestas)
# Cuando las condiciones se cumplan, se pararÃ¡ el juego con un break
while True:
    # Hacemos esta comprobaciÃ³n para parar el bucle principal cuando hemos salido del bucle interno si es necesario
    if respuestas_incorrectas == 3 or respuestas_correctas == 5:
        break
    # Generamos una pregunta aleatoria de la lista
    pregunta_random = random.choice(preguntas_respuestas_lista)
    # Llevamos la cuenta de las preguntas hechas, independientemente de si son correctas o no 
    conteo_preguntas += 1
    print(f"Esta es la pregunta nÃºmero {conteo_preguntas}")
    # Reiniciar intentos permitidos para la nueva pregunta
    intentos_pregunta = 2
    # Bucle para intentar responder la misma pregunta hasta agotar los intentos
    while intentos_pregunta > 0:
        # Mostramos pregunta al usuario y recogemos respuesta
        respuesta_usuario = input(f"Responda la siguiente pregunta: {pregunta_random}")
        # Protegemos el formato de la respuesta del usuario para que coincida con la respuesta del diccionario
        respuesta_usuario = "".join(respuesta_usuario.split()).lower()
        # Comparamos la respuesta del usuario con el valor correspondiente a la pregunta (key) del diccionario
        if respuesta_usuario == preguntas_respuestas[pregunta_random]:
            respuestas_correctas += 1
            print(f"Â¡Has acertado! Llevas un total de {respuestas_correctas} respuestas correctas ğŸ¤©")
            if respuestas_correctas == 5:
                print(f"Enhorabuena, has ganado el juego! ğŸ‰")
                break # Sale del bucle interno
            else:
                print(f"A por la siguiente pregunta! ğŸ¤“")
                # Si adivina una pregunta y el juego sigue, sacamos esa pregunta de la lista para que no se repita
                preguntas_respuestas_lista.remove(pregunta_random)
                break  # Sale del bucle interno y pasa a una nueva pregunta
        else:
            intentos_pregunta -= 1
            if intentos_pregunta > 0:
                print(f"Respuesta incorrecta. Te queda {intentos_pregunta} intento. Â¡Vuelve a intentarlo! ğŸ’ª")
            else:
                respuestas_incorrectas += 1
                print(f"Has fallado la misma pregunta dos veces. Tienes {respuestas_incorrectas} respuestas incorrectas. ğŸ™")
                if respuestas_incorrectas == 3:
                    print(f"Has alcanzado el mÃ¡ximo de respuestas incorrectas. El juego ha terminado... ğŸ˜“")
                    break  # Salimos del bucle de intentos y pasamos a la siguiente pregunta
```
