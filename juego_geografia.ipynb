    - Cada pregunta tiene una respuesta Ãºnica y correcta.

    - El jugador tiene un nÃºmero limitado de intentos para responder correctamente a cada pregunta.

    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.

```python
#Diccionario con preguntas y respuestas (8 preguntas) #DUDA: condiciones ok? 8 preguntas, 5 respuestas, 3 fallos

# Importar la librerÃ­a random
import random

# Crear el diccionario con las preguntas y respuestas

preguntas_respuestas = {
    'Â¿CuÃ¡l es la capital de Francia?': 'paris',
    'Â¿CuÃ¡l es el continente mÃ¡s grande?': 'asia',
    'Â¿QuÃ© paÃ­s tiene la mayor poblaciÃ³n del mundo?': 'india',
    'Â¿CuÃ¡l es el rÃ­o mÃ¡s largo del mundo?': 'amazonas',
    'Â¿En quÃ© paÃ­s se encuentra la Torre Eiffel?': 'francia',
    'Â¿CuÃ¡l es la capital de JapÃ³n?': 'tokio',
    'Â¿CuÃ¡l es el desierto caliente mÃ¡s grande del mundo?': 'sahara',
    'Â¿CuÃ¡l es la capital de Argentina?': 'buenosaires',
    'Â¿QuÃ© paÃ­s es conocido como la tierra de los canguros?': 'australia'
}

#PDTE: Borrar este dicc despuÃ©s de las pruebas
"""
preguntas_respuestas = {
    "CuÃ¡l es la capital de Castilla-La Mancha?": "toledo",
    "CuÃ¡l es el desierto mÃ¡s grande del mundo?": "laantartida",
    "CuÃ¡l es el paÃ­s con mÃ¡s poblaciÃ³n del mundo?": "india"
}
"""

#PDTE: Borrar este print 
print(f"El diccionario generado es {preguntas_respuestas}")

# Definir variables 
intentos_pregunta = 0 #PDTE: Como extra: tenemos posibilidad de fallar 3 veces pero solo suma 1 fallo si llego al intento nÂº3 por pregunta
conteo_preguntas = 0
respuestas_correctas = 0
respuestas_incorrectas = 0

# Pasamos el diccionario a lista porque el mÃ©todo random choice se aplica a listas
preguntas_respuestas_lista = list(preguntas_respuestas)
#PDTE: Borrar este print 
print(f"La lista generada del dict es: {preguntas_respuestas_lista}")

# Usamos condiciÃ³n While True porque el juego se lanzarÃ¡ hasta que se cumplan las condiciones indicadas
# Cuando las condiciones se cumplan, se pararÃ¡ el juego con un break
while True: 
    # Generamos una pregunta aleatoria de la lista
    pregunta_random = random.choice(preguntas_respuestas_lista)
    #PDTE: Borrar este print 
    print(f"pregunta random es {pregunta_random}")
    
    # Llevamos la cuenta de las preguntas hechas, independientemente de si son correctas o no 
    conteo_preguntas += 1
    print(f"Esta es la pregunta nÃºmero {conteo_preguntas}")
    
    # Mostramos pregunta al usuario y recogemos respuesta
    respuesta_usuario = input(f"Responda la siguiente pregunta {pregunta_random}")
    # Protegemos el formato de la respuesta del usuario para que coincida con la respuesta del diccionario
    # Con split eliminamos espacios, con join unificamos en un solo string y con lower lo pasamos todo a minÃºscula
    # PDTE: proteger los acentos -> para el final, es complejo -> importar libreria unicode
    respuesta_usuario = "".join(respuesta_usuario.split()).lower()
    #PDTE: Borrar este print 
    print(f"La respuesta del usuario es: {respuesta_usuario}")
 
    # Comparamos la respuesta del usuario con el value correspondiente a la pregunta (key) del diccionario
    # La variable Pregunta_random recoge la key del diccionario 
    # La variable respuesta_usuario recoge el value del diccionario
    # preguntas_respuestas[pregunta_random] nos da el value de la pregunta hecha al usuario -> lo comparamos con la respuesta dada para ver si es correcta
    if respuesta_usuario == preguntas_respuestas[pregunta_random]:
        #PDTE: Borrar este print 
        print(f"respuesta_usuario es {respuesta_usuario} y preguntas_respuestas[pregunta_random] es {preguntas_respuestas[pregunta_random]}")
        respuestas_correctas += 1
        print(f"Has acertado! Llevas un total de {respuestas_correctas} respuestas correctas")
        #PDTE: Actualizar a respuestas_correctas == 5
        if respuestas_correctas == 5:
            # PDTE: Se puede cambiar print y break por un return
            print(f"Enhorabuena, has ganado el juego! ğŸ‰")
            break
        else:
            print(f"A por la siguiente pregunta!ğŸ¤“")
            # Si adivina una pregunta y el juego sigue, sacar esa pregunta de la lista para que no se repita
            # UsarÃ© remove para eliminar la pregunta de la lista porque remove elimina la primera apariciÃ³n del valor concreto y en este caso solo aparece una vez
            # Podemos usar Pregunta_random que es cada uno de los elementos de la lista "preguntas_respuestas_lista" que es de donde salen las preguntas que lanza la mÃ¡quina
            # PDTE: Borrar estos print
            print(f"La lista de posibles preguntas ANTES de actualizarse es: {preguntas_respuestas_lista}")
            preguntas_respuestas_lista.remove(pregunta_random)
            # PDTE: Borrar estos print
            print(f"La pregunta a borrar es: {pregunta_random}")
            # PDTE: Borrar estos print
            print(f"La lista de posibles preguntas DESPUES de actualizarse es: {preguntas_respuestas_lista}")
    else:
        respuestas_incorrectas += 1
        print(f"La respuesta no es correcta ğŸ™ Has fallado un total de {respuestas_incorrectas} veces")
        #PDTE: Actualizar a respuestas_correctas == 5
        if respuestas_incorrectas == 3:
            print(f"El juego ha terminado... ğŸ˜“")
            break
        else:
            print("Pero no pasa, vuelve a intentarlo! ğŸ’ª") # PDTE: Extra: tiene que volver a lanzar la misma pregunta si falla, darle 3 intentos, si falla los 3, le lanzas otra pregunta y le quitas 1 vida
                                                            # Duda: si fallo 3 veces la misma pregunta la elimino como posible pregunta a futuro?
  
        



```
