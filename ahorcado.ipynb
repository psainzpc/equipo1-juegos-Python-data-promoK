```python
import random 
# creamos lista de palabras secretas
palabras_secretas = ["oro", "jupiter", "coche"] 
# Random choice para que el ordenador elija una palabra de manera aleatoria
palabra_secreta_aleatoria = random.choice(palabras_secretas) 
# Contador de vidas restantes (por cada parte del cuerpo)
vidas_usuario = 6
# Encriptamos la palabra secreta con "_"
palabra_oculta = " _ " * len(palabra_secreta_aleatoria) # Dejamos espacios con el gui칩n para que se vea el n칰mero de letras que tiene la palabra
print(f"Palabra oculta: {palabra_oculta}")

while True:
    # Recogemos la letra del usuario
    letra_usuario = input(f"Indica una letra de la {palabra_oculta}").lower()
    print(f"La letra indicada es {letra_usuario}")
    # Protegemos que indique solo 1 letra
    if len(letra_usuario) > 1:
        print(f"Debe ingresar solo 1 letra")
    else:
        if letra_usuario in palabra_secreta_aleatoria: 
            # Creamos lista de la palabra oculta y palabra_secreta_aleatoria para cambiar las letras que adivine el usuario mediente el 칤ndice de ambas listas
            palabra_oculta = palabra_oculta.replace(" ","") # Eliminamos el espacio del string de la palabra oculta para que no los incluya en la lista -> incluimos espacios para que se viera mejor el n칰meror de guiones 
            palabra_oculta_lista = list(palabra_oculta)  
            palabra_secreta_aleatoria_lista = list(palabra_secreta_aleatoria) 
            # Generamos un bucle for que iterar치 por los 칤ndices del rango de la palabra secreta -> i = 0, i = 1 ... as칤 hasta el len de palabra_secreta_aleatoria_lista
            for i in range(len(palabra_secreta_aleatoria_lista)): 
                # Obtenemos el 칤ndice donde est치 la letra del usuario dentro de la palabra secreta
                if palabra_secreta_aleatoria_lista[i] == letra_usuario: 
                    # Sustituimos la letra del usuario en la palabra en la posici칩n del 칤ndice obtenido
                    palabra_oculta_lista[i] = letra_usuario 
            # Paso la palabra oculta lista a string para mostrarla al usuario -> join y usamos como delimitador " ", porque quiero una cadena de texto con separaci칩n entre los guiones
            palabra_oculta = " ".join(palabra_oculta_lista)
            print(f"Palabra oculta es: {palabra_oculta}")
            if "_" not in palabra_oculta:
                print(f"Felicidades, has adivinado la palabra oculta {palabra_oculta} 游꿀")
                break
        else:
            vidas_usuario -= 1
            print(f"La letra {letra_usuario} no est치 en la palabra secreta 游뗴. Te quedan {vidas_usuario} vidas")
            if vidas_usuario == 0:
                print(f"Has agotado todas tus vidas, el juego ha terminado 游. La palabra secreta era {palabra_secreta_aleatoria}")
                break
```
