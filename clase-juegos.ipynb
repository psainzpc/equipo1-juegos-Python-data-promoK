```python
import random
class Juegos_clasicos:
    def __init__(self, nombre):
        self.nombre = nombre
    
    def jugadora(self):
        print(f"Hola {self.nombre}, bienvenida a Juegos Cl√°sicos con Python")
    
    def jugar(self):
        print("¬øA qu√© juego quieres jugar?")
        print("Introduce una de las siguientes opciones:")
        print('''        1 --> Si quieres jugar al AHORCADO
        2 --> si quieres jugar a PIEDRA, PAPEL, TIJERA
        3 --> si quieres jugar a PREGUNTAS Y RESPUESTAS 
        4 --> si quieres jugar al TRES EN RAYA''')
    
        try:
            opcion = int(input("¬øA qu√© juego quieres jugar?: 1, 2, 3 o 4"))      
            if opcion == 1:
                self.ahorcado()
            elif opcion == 2:
                self.p_p_t()
            elif opcion == 3:
                self.geografia()
            elif opcion == 4:
                self.tres_en_raya()
            else:
                print("opcion no v√°lida, introduce un n√∫mero del 1 al 4")
        except ValueError:
            print("formato incorrecto, introduce un n√∫mero del 1 al 4")

    def ahorcado(self):
        print(f"{self.nombre}, ¬øpreparada para jugar al Ahorcado?")
        print("-------------------------------------")
        print("Antes de empezar, te explico las Reglas del Juego")
        print("-------------------------------------")
        print('''Reglas del juego del "Ahorcado":

        - La jugadora intenta adivinar letras para completar la palabra secreta elegida aleatoriamente.

        - Si la jugadora adivina una letra correctamente, se revela en su lugar correspondiente.

        - Si la jugadora adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

        - El objetivo es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.''')

        a = '''
        +---+
        |   |
            |
            |
            |
            |
            |
        ========='''

        b = '''
            +---+
            |   |
            O   |
                |
                |
                |
                |
            ========='''

        c = '''
            +---+
            |   |
            O   |
            |   |
            |   |
                |
                |
            ========='''

        d = '''
            +---+
            |   |
            O   |
           /|   |
            |   |
                |
                |
            ========='''

        e = '''
            +---+
            |   |
            O   |
           /|\  |
            |   |
                |
                |
            ========='''
        f = '''
            +---+
            |   |
            O   |
           /|\  |
            |   |
           /    |
                |
            ========='''

        g = '''
            +---+
            |   |
            O   |
          ===== |
           /|\  |
            |   |
           / \  |
                |
            ========='''

        import random #seleccionar palabras aleatorias de la lista
        import warnings 
        warnings.filterwarnings("ignore", category=SyntaxWarning) #ignorar advertencia en el c√≥digo al mostrar el mu√±equito del ahorcado

        # creamos lista de palabras secretas
        palabras_secretas = ["oro", "jupiter", "coche"] 
        # Random choice para que el ordenador elija una palabra de manera aleatoria
        palabra_secreta_aleatoria = random.choice(palabras_secretas) 
        # Contador de vidas restantes (por cada parte del cuerpo)
        vidas_usuario = 6
        # Encriptamos la palabra secreta con "_"
        palabra_oculta = " _ " * len(palabra_secreta_aleatoria) # Dejamos espacios con el gui√≥n para que se vea el n√∫mero de letras que tiene la palabra
        print(f"Palabra oculta: {palabra_oculta}")
        print('Empieza el juego!!!!!')
        print(a)

        while True:
            # Recogemos la letra del usuario
            letra_usuario = input(f"Indica una letra de la palabra oculta: {palabra_oculta}").lower()
            print('********************************************')
            print(f"La letra indicada es {letra_usuario}")
            # Verificar si el usuario ingres√≥ un n√∫mero o un car√°cter que no sea una letra
            if not letra_usuario.isalpha():
                print("Debe ingresar una letra, no un n√∫mero u otro car√°cter.")
                continue  # Reinicia para que vuelva a intentar
            # Protegemos que indique solo 1 letra
            if len(letra_usuario) > 1:
                print(f"Debe ingresar solo 1 letra")
            else:
                if letra_usuario in palabra_secreta_aleatoria: 
                    # Creamos lista de la palabra oculta y palabra_secreta_aleatoria para cambiar las letras que adivine el usuario mediente el √≠ndice de ambas listas
                    palabra_oculta = palabra_oculta.replace(" ","") # Eliminamos el espacio del string de la palabra oculta para que no los incluya en la lista -> incluimos espacios para que se viera mejor el n√∫meror de guiones 
                    palabra_oculta_lista = list(palabra_oculta)  
                    palabra_secreta_aleatoria_lista = list(palabra_secreta_aleatoria) 
                    # Generamos un bucle for que iterar√° por los √≠ndices del rango de la palabra secreta -> i = 0, i = 1 ... as√≠ hasta el len de palabra_secreta_aleatoria_lista
                    for i in range(len(palabra_secreta_aleatoria_lista)): 
                        # Obtenemos el √≠ndice donde est√° la letra del usuario dentro de la palabra secreta
                        if palabra_secreta_aleatoria_lista[i] == letra_usuario: 
                            # Sustituimos la letra del usuario en la palabra en la posici√≥n del √≠ndice obtenido
                            palabra_oculta_lista[i] = letra_usuario 
                    # Paso la palabra oculta lista a string para mostrarla al usuario -> join y usamos como delimitador " ", porque quiero una cadena de texto con separaci√≥n entre los guiones
                    palabra_oculta = " ".join(palabra_oculta_lista)
                    print(f"Palabra oculta es: {palabra_oculta}")
                    if "_" not in palabra_oculta:
                        print(f"Felicidades, has adivinado la palabra oculta: {palabra_secreta_aleatoria} üéâ")
                        break
                else:
                    vidas_usuario -= 1
                    if vidas_usuario == 0:
                        print(g)
                        print(f"Has agotado todas tus vidas, el juego ha terminado üíÄ. La palabra secreta era: {palabra_secreta_aleatoria}")
                        break
                    
                    elif vidas_usuario == 5:
                        print(b)
                    elif vidas_usuario == 4:
                        print(c)
                    elif vidas_usuario == 3:
                        print(d)
                    elif vidas_usuario == 2:
                        print(e)
                    elif vidas_usuario == 1:
                        print(f)
                    print(f"La letra {letra_usuario} no est√° en la palabra oculta üôÅ: {palabra_oculta} . Te quedan {vidas_usuario} vidas")
            
                    
                    

    def p_p_t(self):

        print(f"Hola {self.nombre}, ¬øpreparada para jugar a PIEDRA, PAPEL, TIJERA?")
        print("-------------------------------------")
        print("Antes de empezar, te explico las Reglas del Juego")
        print("-------------------------------------")
        print('''Reglas del juego del Piedra, Papel, Tijera:
              
              - La jugadora se enfrentar√° contra el ordenador.
              - La jugadora elige una de las tres opciones: "Piedra", "Papel" o "Tijera". 
              - Las reglas para determinar a la ganadora son:
                    * Piedra vence a Tijera.
                    * Tijera vence a Papel.
                    * Papel vence a Piedra.
              - La jugadora que gana una ronda obtiene un punto.
              - El juego contin√∫a hasta que una jugadora haya acumulado 3 puntos, convirti√©ndose en la ganadora. ''')
        
        # Definir puntos objetivos mediante variable y hacer otras dos variables para contabilizar los puntos del usuario y del ordenador: 

        obj_puntos = 3 
        puntos_usuario = 0
        puntos_ordenador = 0

        # Preguntar al ordenador qu√© opcion elige mediante random choice de la lista de variable opciones:
        opciones = ['piedra', 'papel', 'tijera']
        respuesta_ordenador = random.choice(opciones)

        while True: 

            # Preguntar al jugador mediante input cu√°l es la opci√≥n que elige:
            respuesta_usuario = input('Escoge una opci√≥n: piedra, papel o tijera').lower()
            
            if respuesta_usuario not in opciones:
                print('Introduce una de las opciones')

                respuesta_usuario = input('Escoge una opci√≥n v√°lida: piedra, papel o tijera')  

            elif respuesta_usuario == respuesta_ordenador: 
                
                print(f'Tu respuesta: {respuesta_usuario}')
                print(f'Respuesta ordenador: {respuesta_ordenador}')
                print('Empate')
                print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
                print('---------------------------------')
                
            elif respuesta_usuario == 'piedra' and respuesta_ordenador == 'papel' or respuesta_usuario == 'papel' and respuesta_ordenador == 'tijera' or respuesta_usuario == 'tijera' and respuesta_ordenador == 'piedra':
                puntos_ordenador += 1
                print(f'Tu respuesta: {respuesta_usuario}')
                print(f'Respuesta ordenador: {respuesta_ordenador}')
                print('Has perdido esta ronda')
                print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
                print('---------------------------------')
                
                if puntos_ordenador == obj_puntos:
                    print('Has perdido üòû')
                    break
            elif respuesta_usuario == 'piedra' and respuesta_ordenador == 'tijera' or respuesta_usuario == 'papel' and respuesta_ordenador == 'piedra' or respuesta_usuario == 'tijera' and respuesta_ordenador == 'papel':
                puntos_usuario += 1
                print(f'Tu respuesta: {respuesta_usuario}')
                print(f'Respuesta ordenador: {respuesta_ordenador}')
                print('Has ganado esta ronda')
                print(f'Marcador: usuario: {puntos_usuario} - ordenador: {puntos_ordenador}')
                print('---------------------------------')
                
                if puntos_usuario == obj_puntos: 
                    print('¬°Felicitaciones! Has ganado! üéâ')
                    break
    

        
    
    def geografia(self):
        print(f"Hola {self.nombre}, pon a pruebas tus conocimientos sobre Geograf√≠a con el juego de Preguntas y Respuestas")
        print("-------------------------------------")
        print("Antes de empezar, te explico las Reglas del Juego")
        print("-------------------------------------")
        print('''Reglas del Juego de "Preguntas y Respuestas de Geograf√≠a":
              
    - El juego consiste en adivinar la respuesta correcta sobre preguntas de geograf√≠a.
              
    - La jugadora tendr√° que introducir la respuesta que considere correcta.

    - Cada pregunta tiene una respuesta √∫nica y correcta.

    - La jugadora tiene 2 intentos para responder correctamente a cada pregunta.

    - El juego sigue hasta que la jugadora responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.''')
    

        # Diccionario con preguntas y respuestas
        preguntas_respuestas = {
            '¬øCu√°l es la capital de Francia?': 'paris',
            '¬øCu√°l es el continente m√°s grande?': 'asia',
            '¬øQu√© pa√≠s tiene la mayor poblaci√≥n del mundo?': 'india',
            '¬øCu√°l es el r√≠o m√°s largo del mundo?': 'amazonas',
            '¬øEn qu√© pa√≠s se encuentra la Torre Eiffel?': 'francia',
            '¬øCu√°l es la capital de Jap√≥n?': 'tokio',
            '¬øCu√°l es el desierto caliente m√°s grande del mundo?': 'sahara',
            '¬øCu√°l es la capital de Argentina?': 'buenosaires',
            '¬øQu√© pa√≠s es conocido como la tierra de los canguros?': 'australia'
        }

        # Variables
        intentos_pregunta = 2  # Intentos permitidos por cada pregunta
        conteo_preguntas = 0
        respuestas_correctas = 0
        respuestas_incorrectas = 0

        # Pasamos el diccionario a lista porque el m√©todo random choice se aplica a listas
        preguntas_respuestas_lista = list(preguntas_respuestas)
        # Cuando las condiciones se cumplan, se parar√° el juego con un break
        while True:
            # Hacemos esta comprobaci√≥n para parar el bucle principal cuando hemos salido del bucle interno si es necesario
            if respuestas_incorrectas == 3 or respuestas_correctas == 5:
                break
            # Generamos una pregunta aleatoria de la lista
            pregunta_random = random.choice(preguntas_respuestas_lista)
            # Llevamos la cuenta de las preguntas hechas, independientemente de si son correctas o no 
            conteo_preguntas += 1
            print(f"Esta es la pregunta n√∫mero {conteo_preguntas}")
            # Reiniciar intentos permitidos para la nueva pregunta
            intentos_pregunta = 2
            # Bucle para intentar responder la misma pregunta hasta agotar los intentos
            while intentos_pregunta > 0:
                # Mostramos pregunta al usuario y recogemos respuesta
                respuesta_usuario = input(f"Responda la siguiente pregunta: {pregunta_random}")
                # Protegemos el formato de la respuesta del usuario para que coincida con la respuesta del diccionario
                respuesta_usuario = "".join(respuesta_usuario.split()).lower()
                # Comparamos la respuesta del usuario con el valor correspondiente a la pregunta (key) del diccionario
                if respuesta_usuario == preguntas_respuestas[pregunta_random]:
                    respuestas_correctas += 1
                    print(f"¬°Has acertado! Llevas un total de {respuestas_correctas} respuestas correctas ü§©")
                    if respuestas_correctas == 5:
                        print(f"Enhorabuena, has ganado el juego! üéâ")
                        break # Sale del bucle interno
                    else:
                        print(f"A por la siguiente pregunta! ü§ì")
                        # Si adivina una pregunta y el juego sigue, sacamos esa pregunta de la lista para que no se repita
                        preguntas_respuestas_lista.remove(pregunta_random)
                        break  # Sale del bucle interno y pasa a una nueva pregunta
                else:
                    intentos_pregunta -= 1
                    if intentos_pregunta > 0:
                        print(f"Respuesta incorrecta. Te queda {intentos_pregunta} intento. ¬°Vuelve a intentarlo! üí™")
                    else:
                        respuestas_incorrectas += 1
                        print(f"Has fallado la misma pregunta dos veces. Tienes {respuestas_incorrectas} respuestas incorrectas. üôÅ")
                        if respuestas_incorrectas == 3:
                            print(f"Has alcanzado el m√°ximo de respuestas incorrectas. El juego ha terminado... üòì")
                            break  # Salimos del bucle de intentos y pasamos a la siguiente pregunta
 
    def tres_en_raya(self):
        print(f"Hola {self.nombre}, ¬øpreparada para jugar al cl√°sico TRES EN RAYA?")
        print("-------------------------------------")
        print("Antes de empezar, te explico las Reglas del Juego")
        print("-------------------------------------")
        print('''Reglas del juego Tres en Raya:
              - El juego se juega en un tablero de 3x3.
              - Dos jugadores, uno representa "X" y el otro "O", alternan turnos.
              - El objetivo es obtener tres s√≠mbolos del mismo tipo en l√≠nea horizontal, vertical o diagonal.
              - El juego termina cuando una jugadora gana o el tablero se llena (empate).
              - La jugadora que logre colocar tres de sus s√≠mbolos en l√≠nea es el ganador.
 ''')

        # Tablero inicial
        huecos_tablero = {'a1':'-', 'b1':'-', 'c1':'-', 'a2':'-', 'b2':'-', 'c2':'-', 'a3':'-', 'b3':'-', 'c3':'-'} 
        #Lista de coordenadas v√°lidas coordenadas_validas = ['a1', 'b1', 'c1', 'a2', 'b2', 'c2', 'a3', 'b3', 'c3']
        # Funci√≥n para imprimir el tablero
        def imprimir_tablero(huecos_tablero):
            tablero = f'''
            +-----+-----+-----+
            |  a  |  b  |  c  |
        _______________________
        | 1 |  {huecos_tablero['a1']}  |  {huecos_tablero['b1']}  |  {huecos_tablero['c1']}  |
        _______________________
        | 2 |  {huecos_tablero['a2']}  |  {huecos_tablero['b2']}  |  {huecos_tablero['c2']}  |
        _______________________
        | 3 |  {huecos_tablero['a3']}  |  {huecos_tablero['b3']}  |  {huecos_tablero['c3']}  |
        +---+-----+-----+-----+
        '''
            print(tablero)
            
        # Funci√≥n para comprobar si hay un ganador o empate
        def comprobar_ganador(huecos_tablero):
            # Combinaciones ganadoras
            combinaciones_ganadoras = [
                ['a1', 'b1', 'c1'], ['a2', 'b2', 'c2'], ['a3', 'b3', 'c3'],  # Filas
                ['a1', 'a2', 'a3'], ['b1', 'b2', 'b3'], ['c1', 'c2', 'c3'],  # Columnas
                ['a1', 'b2', 'c3'], ['a3', 'b2', 'c1']  # Diagonales
            ]
            # Revisamos si alguna combinaci√≥n tiene el mismo valor (X o O)
            for combo in combinaciones_ganadoras:
                if huecos_tablero[combo[0]] == huecos_tablero[combo[1]] == huecos_tablero[combo[2]] and huecos_tablero[combo[0]] != '-':
                    return huecos_tablero[combo[0]]  # Retornamos 'X' o 'O' seg√∫n quien gan√≥
            # Comprobar si hay empate
            if '-' not in huecos_tablero.values():
                return 'E'  # Empate
            return # No hay ganador ni empate si devuelve None   

        # Funcion principal del juego
        #def juego_tic_tac_toe():
        while True:
            imprimir_tablero(huecos_tablero)
            
            # Comprobamos si hay ganador o empate antes de volver a iniciar la jugada
            resultado = comprobar_ganador(huecos_tablero)
            if resultado == "X":
                print('Ha ganado el jugador X! Enhorabuena üéâüéâüéâ')
                break
            elif resultado == "O":
                print('Ha ganado el jugador O! Enhorabuena üéâüéâüéâ')
                break
            elif resultado == "E":
                print('Empate! üòÆ')
                break

            # Juega X 
            coordenada_X = (input('JUGADOR X, Dime la coordenada para tu ficha')).lower()
            print(f"Jugador X: {coordenada_X}")
            if huecos_tablero[coordenada_X] == '-': #comprobando que el hueco est√° disponible
                print('El hueco est√° disponible')#si lo est√°, cambia '-' por X
                huecos_tablero[coordenada_X]= 'X'
                imprimir_tablero(huecos_tablero)
                # Comprobamos si hay ganador o empate antes de volver a iniciar la jugada
                resultado = comprobar_ganador(huecos_tablero)
                if resultado == "X":
                    print('Ha ganado el jugador X! Enhorabuena üéâüéâüéâ')
                    break
                elif resultado == "E":
                    print('Empate! üòÆ')
                    break             
            else:    
                print ('Ese hueco no est√° disponible. Indique otro por favor')
            
            # Hacemos un bucle interno para dar la posibilidad al jugador O de volver a meter una coordenada en caso de que se equivoque    
            while True:
                # Juega O
                coordenada_O = (input('JUGADOR O, Dime la coordenada para tu ficha').lower())
                print(f"Jugador O: {coordenada_O}")
                if huecos_tablero[coordenada_O] == '-':
                    print('El hueco est√° disponible')
                    huecos_tablero[coordenada_O] = 'O'
                    # Salimos del bucle interno para evaluar si ha ganado el jugador O, si hay empate o si el juego sigue, dando en este caso el turno al jugador X
                    break
                else:
                    print ('Ese hueco no est√° disponible. Indique otro por favor')


    
```
